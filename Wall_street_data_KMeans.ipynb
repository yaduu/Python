```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('wall_street_data.csv')
```

```python
data.head()
```

```python
del data['Unnamed: 0']
```

```python
samples=data.values
```

```python
samples
```

```python
from sklearn import preprocessing
```

```python
normalized= preprocessing.normalize(samples)
```

```python
normalized
```

```python
ks=range(1,60)
inertias=[]

for k in ks:
    model=KMeans(n_clusters=k)
    model.fit(normalized)
    inertias.append(model.inertia_)
```

```python
inertias
```

```python
sns.lineplot(range(1,60),inertias)
```

```python
model=KMeans(n_clusters=10)
```

```python
labels=model.fit_predict(samples)
```

```python
labels
```

```python
df=pd.DataFrame({'labels':labels,'brand':data['Unnamed: 0']})
```

```python
df.sort_values(ascending=True,by='labels')
```
